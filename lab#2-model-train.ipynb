```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
dataset = pd.read_csv("data.csv")
```

```python
dataset.head()
```

```python
# Remove unwanted features.

del dataset['id']
del dataset['Unnamed: 32']
dataset.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(dataset.loc[:, dataset.columns != 'diagnosis'],
                                                    dataset['diagnosis'], 
                                                    stratify=dataset['diagnosis'], 
                                                    random_state=0)
```

```python
model = LogisticRegression().fit(X_train, y_train)

print(f"Logistic regression training set classification score: {format(model.score(X_train, y_train), '.4f')} ")
print(f"Logistic regression testing set classification score: {format(model.score(X_test, y_test), '.4f')} ")
```

```python
# Save to file in the current working directory
import pickle

pkl_filename = "cancer_model.pkl"

with open(pkl_filename, 'wb') as file:
    pickle.dump(model, file)
```
